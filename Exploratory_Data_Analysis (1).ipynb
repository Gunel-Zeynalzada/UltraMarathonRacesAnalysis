```python
#Clean up athlete age
df2['athlete_age']= 2020 - df2['Athlete year of birth']
```

```python
#drop columns
df2=df2.drop(['Athlete club','Athlete country','Athlete year of birth','Athlete age category'], axis=1)
```

```python
df2=df2.dropna()
```

```python
# Rename columns
df2 = df2.rename(columns={
    'Year of event': 'year',
    'Event dates': 'race_date',
    'Event name': 'race_name',
    'Event distance/length': 'race_distance',
    'Event number of finishers': 'race_number_of_finishers',
    'Athlete performance': 'athlete_performance',
    'Athlete gender': 'athlete_gender',
    'Athlete average speed': 'athlete_average_speed',
    'Athlete ID': 'athlete_id'
})
```

```python
df2['athlete_age']=df2['athlete_age'].astype(int)
df2['athlete_average_speed']=df2['athlete_average_speed'].astype(float)
```

```python
df2.head()
```

```python
#Charts and graphs
sns.histplot(df2['race_distance'])
```

```python
sns.histplot(df2, x='race_distance', hue='athlete_gender')
```

```python
sns.displot(df2[df2['race_distance']=='50mi']['athlete_average_speed'])
```

```python
sns.violinplot(data=df2, x='race_distance', y='athlete_average_speed', hue='athlete_gender', split=True, inner='quart', linewidth=1)
```

```python
sns.lmplot(data=df2, x='athlete_age', y='athlete_average_speed', hue='athlete_gender')
```

```python
#Difference in speed for male and female
df2.groupby(['race_distance','athlete_gender'])['athlete_average_speed'].mean()
```

```python
#What age groups are the worst in the 50mi Race(20 + races min) (show 15)
df2.query('race_distance=="50mi"').groupby(['athlete_age'])['athlete_average_speed'].agg(['mean','count']).sort_values('mean', ascending=True).query('count>19').head(15)
```

```python
#What age groups are the best in the 50mi Race(20 + races min) (show 15)
df2.query('race_distance=="50mi"').groupby(['athlete_age'])['athlete_average_speed'].agg(['mean','count']).sort_values('mean', ascending=False).query('count>19').head(15)
```

```python
df2['race_month']=df2['race_date'].str.split('.').str.get(1).astype(int)
```

```python
df2['race_season']=df2['race_month'].apply(lambda x: 'Spring' if 3<=x<=5 else 'Summer' if 6<=x<=8 else 'Fall' if 9<=x<=11 else 'Winter')
```

```python
df2.groupby('race_season')['athlete_average_speed'].agg(['mean','count']).sort_values('mean', ascending=False)
```
